Создаем 10к документов для кластерного анализа:

```python
#в этой переменной будут слова через проблем для одного документа потом запятая и другой дкоумент [я ты он, они я ты]
vw_data = []
document = []
k=0
pattern = re.compile('_(?:PREP|X|ADJ_PRONOUN|PUNCT|NOUN_PRONOUN|CONJ|PART|NUM|ADV|ADV_PRONOUN|ADVPARTICIPLE|NN_NOUN)|http|[^a-zа-яА-ЯA-Z_]')
with open("C:\DB\kurgansky_lemmas_0.txt", encoding="utf-8-sig") as f:
    for i,doc in enumerate(f):
        if k>9999:
            break
        for word in doc.split():
            if len(re.sub('_.*','', word)) > 2:
                if word == "########" and i>0:
                    vw_data.append(' '.join(document))
                    document = []
                    k +=1
                elif not pattern.search(word) and i>0:
                    document.append(word)
```

Вот отсюда кластерный анализ начинаем:

```python
#в этой переменной просто сохраняем метод, который из нашего списка документов и слов будет делать матрицу
vectorizer = TfidfVectorizer()
#А вот тут мы получаем как раз эту матрицу применяем метод "векторизировать" к нашим документам с текстами
X = vectorizer.fit_transform(vw_data)
```

```python
true_k = 50
#в этой переменной тупо храним модель и настройки
model = KMeans(n_clusters=true_k, init='k-means++', max_iter=100, n_init=1)
#Потом эту модель применяем на данных Х (матрице документ на слово, сильно разряженной так как не каждое слово в каждом документе)
model.fit(X)
```

Сохраняем кластеры в список:

```python
list_with_terms = []
order_centroids = model.cluster_centers_.argsort()[:, ::-1]
terms = vectorizer.get_feature_names()
for i in range(true_k):
    list_with_terms.append([])
    for ind in order_centroids[i, :10]:
        list_with_terms[i].append(terms[ind])
```

```python
documents_with_marked_terms = []
number_of_document = -1
find_stop_word_flag = 0

with open("C:\DB\kurgansky_lemmas_0.txt", encoding="utf-8-sig") as f:
    for paragraph in f:
        for word in paragraph.split():
            if len(re.sub('_.*','', word)) > 2:
                if word == "########":
                    documents_with_marked_terms.append([])
                    number_of_document += 1
                else:
                    for i in ['_PREP', '_X', '_ADJ_PRONOUN', '_PUNCT', '_NOUN_PRONOUN', '_CONJ', '_PART', 'NUM', 'ADV', 'ADV_PRONOUN', 'ADVPARTICIPLE']:
                        pattern = re.compile(i)
                        if pattern.search(word):
                            find_stop_word_flag = 1
                            break
                    if find_stop_word_flag != 1:
                        documents_with_marked_terms[number_of_document].append(word)
                    find_stop_word_flag = 0
```

```python
Documents_with_separeted_words = []
for i in range(len(documents_with_marked_terms)):
    Documents_with_separeted_words.append(', '.join(documents_with_marked_terms[i]))
```

```python
vocab = []
docword = []

for i, document in enumerate(documents_with_marked_terms):
    if i > 9999:
        break
    docword.append([])
    for word in document:
        try:
            index_value = vocab.index(word) #в этом случае индекс минус еденичный
        #Если не находим в словаре, то
        except ValueError:
            #добавляем слово в словарь
            index_value = len(vocab)
            vocab.append(word) 
        flag = 0
        for j, w in enumerate(docword[i]):
            if w[1] == index_value+1:
                flag = 1
                break
        if flag == 1:
            docword[i][j][2] += 1
        else:
            docword[i].append([i+1,index_value+1,1])
        flag = 0
```
